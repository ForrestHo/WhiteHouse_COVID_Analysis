```python
# Generic function to get links, check for key words and append to main list

from bs4 import BeautifulSoup
import requests

key_words = ['CORONAVIRUS', 'Coronavirus', 'coronavirus', 'Covid-19', 'COVID-19', 'COVID', 'covid-19', 'SARS-CoV-2']
all_links = []

def find_covid(link):
    source = requests.get(link).text
    soup = BeautifulSoup(source, 'lxml')
    
    for article in soup.find_all('div', class_='briefing-statement__content'):
        link = article.h2.a['href']
        for word in key_words:
            if word in link:
                all_links.append(link)
```

```python
# Scrapping all dates, titles, articles and links

from bs4 import BeautifulSoup
import requests
import json

with open('all_links.json') as json_file:
    all_links = json.load(json_file)


all_dates = []
all_titles = []
all_articles = []

for link in all_links:
    
    source = requests.get(link).text
    soup = BeautifulSoup(source, 'lxml')
    
    date = soup.find('p', class_='meta__date').time.text
    all_dates.append(date)
    
    title = soup.find('div', class_='page-header__wrap').h1.text
    all_titles.append(title)
    
    text = soup.find('div', class_='page-content__content editor').text
    all_articles.append(text)
```

```python
# Create clean text function
import re
import string

def clean_text_1(text):
    text = text.lower()
    text = re.sub(r'\n|\t|\xa0', ' ', text)
    text = re.sub(r'share:|share this page on facebook|share this page on twitter|copy url to your clipboard', '', text)
    text = re.sub(r'all news|quote', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.translate(str.maketrans('', '', string.digits))
    return text

round1 = lambda x: clean_text_1(x)
```
